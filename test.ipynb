```python
import folium
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "col1": np.random.randn(100) / 500 + 34.81732,
    "col2": np.random.randn(100) / 500 + 113.535807,
    "col3": np.random.randn(100) * 10,
    "col4": np.random.rand(100, 4).tolist(),
})

# 创建地图对象
m = folium.Map(location=[34.81732, 113.535807], zoom_start=14)
folium.TileLayer('Stamen Terrain').add_to(m)  # 使用Stamen Terrain提供商的影像地图
# 将DataFrame数据添加到地图上
for index, row in df.iterrows():
    folium.CircleMarker(
        location=[row['col1'], row['col2']],
        radius=row['col3'],
        color='#3186cc',
        fill=True,
        fill_color='#3186cc'
    ).add_to(m)

# 在notebook中显示地图
m
```
